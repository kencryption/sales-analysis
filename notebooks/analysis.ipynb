# Superstore Sales & Profit Analysis

This analysis explores the Superstore dataset to understand
sales patterns and profitability.

Goal:
Identify the key profit drivers and uncover business problems.

### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/superstore.csv", encoding="latin1")
df.head()
```

## Dataset Overview

```python
df.shape
```

```python
df.info()
```

## Statistical Summary

```python
df.describe()
```

## Data Preprocessing

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

Conversion Of Date Columns

```python
df.info()
```

## Feature Engineering

```python
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

## Overview Of Superstore

Total Revenue

```python
df['Sales'].sum()
```

Total Profit

```python
df['Profit'].sum()
```

Average Discount

```python
print(f"{df['Discount'].mean() * 100:.2f}%")
```

## Sales By Category

```python
sales_by_category = df.groupby("Category")["Sales"].sum()
```

```python
sns.barplot(x=sales_by_category.index, y=sales_by_category.values)
plt.title("Sales by Category")
plt.savefig("../outputs/charts/sales_by_category.png")
plt.show()
```

### Insight
Technology generates the highest sales. Second highest sales generating category is Furniture.

## Profit By Sub-Category

```python
profit_by_sub_category = df.groupby("Sub-Category")["Profit"].sum().sort_values()
```

```python
profit_by_sub_category.plot(kind="barh", figsize=(8,6))
plt.title("Profit by Sub-Category")
plt.savefig("../outputs/charts/profit_by_subcategory.png")
plt.show()
```

### Insight
Copiers generate the highest profit whereas tables generate significant losses compared to other sub-categories.

## Sales vs Profit Relationship

```python
sns.scatterplot(data=df, x="Sales", y="Profit")
plt.title("Sales vs Profit")
plt.savefig("../outputs/charts/sales_vs_profit.png")
plt.show()
```

### Insight
Sales and profit show a positive relationship showing that higher sales generate higher profit, however, a few high sales orders result in negative profit suggesting heavy discounts.

## Sales By Customer Segment

```python
sns.barplot(data=df, x="Segment", y="Sales")
plt.title("Sales by Customer Segment")
plt.savefig("../outputs/charts/sales_by_customer_segment.png")
plt.show()
```

### Insight
Home Office customers generates the highest sales followed by Corporate and Consumer segments.

## Loss Orders

```python
loss_orders = df[df["Profit"] < 0]

loss_orders.shape[0]
```

```python
sns.histplot(loss_orders["Discount"])
plt.title("Discount Distribution for Loss Orders")
plt.savefig("../outputs/charts/loss_orders_discount.png")
```

### Insight
Loss-making orders are strongly associated with higher discount levels.

## Shipping Mode vs Shipping Time

```python
sns.boxplot(data=df, x="Ship Mode", y="Shipping Time")
plt.title("Shipping Mode vs Ship Time")
plt.savefig("../outputs/charts/ship_mode_vs_ship_time.png")
plt.show()
```

### Insight
Same Day deliveries are always fast whereas Standard Class shipments take the longest compared to other shipping modes.
