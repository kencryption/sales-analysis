# Feature Engineering for Profit Prediction

Now we prepare the dataset for machine learning by creating new features and transforming categorical variables.

Goal: Construct a dataset suitable for predicting profit and optimizing discount strategies.

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
df = pd.read_csv("../data/raw/superstore.csv", encoding="latin1")
```

## Conversion Of Date Columns

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

## Create Time Features

```python
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df["Day Of Week"] = df["Order Date"].dt.dayofweek
```

## Pricing Features

```python
df["Item Price"] = df["Sales"] / df["Quantity"]
```

```python
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

## Order Size Features

```python
df["Value Per Item"] = df["Sales"] / df["Quantity"]
```

```python
df["Discount Amount"] = df["Sales"] * df["Discount"]
```
